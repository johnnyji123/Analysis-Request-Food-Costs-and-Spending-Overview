```python
import pandas as pd
!pip install polars
import polars as pl
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_excel(r"C:\Users\asus\Documents\Candidate_sample- IT analyst -v2.xlsx", sheet_name = "2.groceries-by prod-6 months",
                   dtype = {
                        "Product Code":str, 
                       "Product Description": str,
                       "Pack size":str,
                       "Major Category": str,
                       "Minor Category": str
                       
# Converting excel file into a pandas dataframe                   }
                  )
df_polars = pl.from_pandas(df)
df_polars

# converting the pandas dataframe into a polars dataframe for data manipulation/analysis and cleaning
```

```python
# Create a chart listing the top 50 products ranked by their total expenditure and volume.
df_polars = df_polars.select(pl.col("Product Code"),
                 pl.col("Product Description"),
                 pl.col("6 months volumes"),
                 pl.col("6 months spend")
                )

df_polars

# Creating a dataframe with the relevant information/data
```

```python
# product by volume 
product_by_volume = df_polars.sort("6 months volumes", descending = True).head(50)
product_by_volume = product_by_volume.with_columns(
    pl.arange(0, 50).alias("Rank by volume")  
)
product_by_volume = product_by_volume.filter(pl.col("Product Description").is_not_null())
product_by_volume

# Found the top 50 products by volume 
# Created an extra column showing the rank of these products from 1 -50
```

```python
# product by expenditure
product_by_expenditure = df_polars.sort("6 months spend", descending = True).head(50)
product_by_expenditure = product_by_expenditure.with_columns(pl.arange(0,50).alias("Rank by expenditure"))
product_by_expenditure = product_by_expenditure.filter(pl.col("Product Description").is_not_null())
product_by_expenditure

# creating a dataframe to find the top 50 productst by expenditure
```

```python
plt.figure(figsize=(14, 8))  # Increase figure size for better visibility
sns.barplot(y="Product Description", x="6 months volumes", data=product_by_volume, palette="viridis")

# Add titles and labels
plt.title("Top Products by Volume (Last 6 Months)", fontsize=16)
plt.ylabel("Product Description", fontsize=12)
plt.xlabel("6 Months Volumes", fontsize=12)

# Show the plot
plt.tight_layout()  # Adjust layout to make room for the labels
plt.show()

# Graph created using seaborn showing the top products by Volume in the last 6 months
```

```python
plt.figure(figsize=(14, 8))  # Increase figure size for better visibility
sns.barplot(y="Product Description", x="6 months volumes", data=product_by_expenditure, palette="viridis")

# Add titles and labels
plt.title("Top Products by Expenditures (Last 6 Months)", fontsize=16)
plt.ylabel("Product Description", fontsize=12)
plt.xlabel("6 Months Expenditures", fontsize=12)

# Show the plot
plt.tight_layout()  # Adjust layout to make room for the labels
plt.show()

# Graph created using seaborn to show the top products by expenditures in the last 6 months
```

```python
# merge df
merged_df = product_by_volume.join(product_by_expenditure, on = "Product Code", how = "left")
merged_df

# From this we can see that the rank 1 by volume is rank 5 by expenditure, and we can see all of this from one table
# We can make various comparisons like this. E.g. the product rank 2 by volume is rank 21 by expenditure. This can show us correlations between expenditure and volume.
```

```python
new_df = df_polars.select(pl.col("Product Code"),
                 pl.col("Product Description"),
                 pl.col("6 months volumes"),
                 pl.col("6 months spend"),
                pl.col("Major Category"),
                pl.col("Minor Category")
                )

new_df

# Creating a new dataframe to calculate the total expenses of the Major Category
```

```python
sum_of_expenditure_by_major_category = new_df.group_by("Major Category").agg(pl.col("6 months spend").sum().alias("total expenses"))
sum_of_expenditure_by_major_category

#  Using aggregate functions to calculate total expenses for Major category
```

```python
overall_expenses_value = sum_of_expenditure_by_major_category.select(pl.col("total expenses").sum().round(2)).item()
overall_expenses_value

# Caclulating the overall expenses based on the total expenses per Major Category
```

```python
spending_details = sum_of_expenditure_by_major_category.with_columns(
    (pl.col("total expenses") / overall_expenses_value * 100).round(2).alias("percentage contribution to overall expenses")
)

# Display the results
spending_details = spending_details.sort("percentage contribution to overall expenses", descending = True)
spending_details = spending_details.filter(pl.col("Major Category").is_not_null())
spending_details = spending_details.head(50)
spending_details = spending_details.with_columns(pl.arange(1,51).alias("ranking based on percentage contributions"))

spending_details
# The code calculates the percentage contribution of total expenses for each major category.
# This is done by dividing the total expenses of each category by the overall expenses value.
# The result is then multiplied by 100 to express it as a percentage.
# The final value is rounded to two decimal places and labeled as "percentage contribution to overall expenses".
# The spending details are sorted in descending order based on the percentage contribution to overall expenses.
# This allows for quick identification of categories with the highest contributions.
# The results are limited to the top 50 major categories based on their percentage contributions.
# The results are limited to the top 50 major categories based on their percentage contributions.
# This helps to focus on the most significant categories in terms of spending. This helps to focus on the most significant categories in terms of spending.
# Can see baseed on ranking which products generate the most expenses e.g. No major category contribiutes the most to overall expenses
```

```python
# bases, flour, eggs, miilk and cream, vegetables
major_category_lst = ["Bases", "Flour", "Eggs", "Milk & Cream", "Vegetables"]
major_category_best_price = new_df.filter(pl.col("Major Category").is_in(major_category_lst))
major_category_best_price = major_category_best_price.sort("6 months spend", descending = False)
major_category_best = major_category_best_price.head(10)
major_category_best

# A list of major categories is created, specifically focusing on "Bases," "Flour," "Eggs," "Milk & Cream," and "Vegetables." This ensures the analysis is targeted at key product groups.
# The DataFrame is filtered to include only products that fall within the specified major categories.
# This results in a new DataFrame that focuses solely on the relevant products for further analysis.
# The filtered DataFrame is sorted by the "6 months spend" column in ascending order.
# Sorting this way allows for quick identification of the products with the lowest expenditures within each major category.
# The filtered DataFrame is sorted by the "6 months spend" column in ascending order.
# Sorting this way allows for quick identification of the products with the lowest expenditures within each major category.

```

```python
unique_product_count = (
    major_category_best
    .group_by("Major Category")
    .agg(pl.col("Product Description").n_unique().alias("Unique Product Count"))
)

# Display the result
unique_product_count

#The DataFrame containing the best-priced products/cheapest is grouped by the "Major Category."
# Grouping allows for aggregation of data within each category, making it easier to analyze specific metrics.
# This output provides a summary of product diversity within the best-priced selections across the specified categories.
# In conclusion:
# Milk & Cream can be found in the top 10 cheapest products 2 times.
# Bases can be found in the top 10 cheapest products 3 times.
# Vegetables can be found in the top 10 cheapest products 4 times.
# Eggs can be found in the top 10 cheapest products 1 time.
```
